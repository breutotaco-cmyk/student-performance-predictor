```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Create dataset
data = {
    "Study Hours": [2, 3, 4, 5, 6, 7, 8, 9, 10],
    "Sleep Hours": [5, 6, 6, 7, 7, 8, 8, 9, 9],
    "Phone Time": [5, 4, 4, 3, 3, 2, 2, 1, 1],
    "Grade": [60, 65, 70, 75, 80, 85, 90, 95, 98]
}

df = pd.DataFrame(data)

# Split data
X = df[["Study Hours", "Sleep Hours", "Phone Time"]]
y = df["Grade"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
prediction = model.predict([[6, 7, 2]])

print("Predicted Grade:", prediction[0])
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Study Hours"], df["Grade"])
plt.xlabel("Study Hours")
plt.ylabel("Grade")
plt.title("Study Hours vs Grade")
plt.show()
```
